```python
def in_range(n, start, end):
    if n >= start and n <= end:
        return True
    return False

print(in_range(1, 2, 20))
print(in_range(12, 3, 30))
```

```python
def is_perfect(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s == n

print(is_perfect(6))
print(is_perfect(10))
```

```python
def is_palindrome(s):
    i = 0
    a = len(s) - 1

    while i < a:
        if s[i] != s[a]:
            return False
        i += 1
        a -= 1

    return True

print(is_palindrome("Sanvi"))
print(is_palindrome("Shah"))
```
